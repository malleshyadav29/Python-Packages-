```python
# for continuous graph (plot)

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
ax.hist(x, bins=8, linewidth=0.5, edgecolor='black', histtype='barstacked')

ax.set(
    xlim=(0, 8), xticks=np.arange(1, 9),
    ylim=(0, 56), yticks=np.linspace(0, 56, 9)
)

plt.show()
```

#4.Scatter Plot:

#5.Pie chart:
- used to get pie chart

```python
import matplotlib.pyplot as plt

labels = ['Academy', 'IT Services', 'BPO Services']
sizes = [55, 35, 40]

plt.pie(sizes, labels=labels, autopct='%1.2f%%')
plt.title('Pie Chart Example')
plt.show()
```

#6.Area Plot:

```python
x=[1,2,3,4,5]
y1,y2=[10,20,15,25,67], [5,15,10,20,25]

plt.fill_between(x, y1, y2, color='red', alpha=0.4, label='Area 1-2')
plt.plot(x,y1,label='Line1',marker='*')
plt.plot(x,y2,label='Line2',marker='_')

plt.xlabel('x-axis'),plt.ylabel('y-axis'),plt.title('Area Chart')
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1,y2=[10,20,15,25,67], [5,15,10,20,10]

plt.fill_between(x, y1, y2, color='green', alpha=0.2, label='Area b/w 1-2')
plt.plot(x, y1, marker='o', label='Line1')
plt.plot(x, y2, marker='o', label='Line2')

plt.title("Area Chart Example"),plt.xlabel("x-axis"),plt.ylabel("y-axis")

plt.grid(False)
plt.legend()
plt.show()
```

#**Introduction to Figure**

```python
# constrained layout automatically adjusts subplots

fig = plt.figure(figsize=(4,5))
fig.suptitle('Figure')

ax = fig.add_subplot()

ax.set_title('Axes', loc='left', fontstyle='oblique', fontsize='medium')
plt.grid(False)
plt.show()
```

#square,square root,exponential,log functions

```python
fig, a=  plt.subplots(2,2)
import numpy as np
x=np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('Sq')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('Sqrt')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```
