#Dealing with missing data techniques

- there are three techniques to deal with missing values
#1.Dropping missing values:
- one way to deal with missing values is to drop it
- this technique is used in huge amount of data because - - losing small amount of data in huge amount will not make impact
- we will use .dropna() method
- Disadvantage :

-       Here we loose so much of data so this is the main disadvantage of this dropna() method


```python
df=pd.read_csv(r"/content/StudentPerformanceFactors.csv")
```

```python
df.shape
```

```python
df.head(12)
```

```python
a=df.dropna(inplace=True, axis=0)
```

```python
print(df.isnull().sum())
```

```python
print(df.shape)
```

#2. Replacing the null value
- in this replace method we use some other value to replace the null value
- here we use mean and median of the data and then replace them


```python
df=pd.read_csv(r"/content/StudentPerformanceFactors.csv" , usecols=['Hours_Studied'])
```

```python
df.isnull().sum()
```

```python
meanvalue=df['Hours_Studied'].mean()        # using mean value
print(meanvalue)
```

```python
data=df.fillna(meanvalue)
```

```python
data.head(12)
```

```python
medianvalue=df['Hours_Studied'].median()  # using medium value
print(medianvalue)
```

```python
data=df.fillna(medianvalue)
```

```python
data.head(12)
```

#Which is better to use mean or median
- It is better to use median instead of mean
- because of the outlayers in the data
- outlayers: data containing out of the range

#Aggregating data by .groupby()
- you can aggregate the data by combining the .groupby() method in summary method
- when you need to clean the data

```python
data=pd.read_csv(r"/content/StudentPerformanceFactors.csv",usecols=['Sleep_Hours','School_Type','Hours_Studied','Physical_Activity'])
```

```python
data.head()
```

```python
meanvalue=data['Sleep_Hours'].mean()
print(meanvalue)
```

```python
data.fillna(meanvalue).head()
```

```python
data.groupby('School_Type').mean().head()
```

```python
data.groupby('School_Type').median().head()
```
